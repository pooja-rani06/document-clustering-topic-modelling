## 📂 Step 2: Load Dataset

```python
# Load 20 Newsgroups dataset
# Using a subset of 6 categories for clarity — feel free to expand
categories = [
    'rec.sport.hockey',
    'sci.space',
    'talk.politics.guns',
    'comp.graphics',
    'rec.motorcycles',
    'sci.med'
]

newsgroups = fetch_20newsgroups(
    subset='all',
    categories=categories,
    remove=('headers', 'footers', 'quotes'),  # Remove metadata noise
    random_state=42
)

documents = newsgroups.data
true_labels = newsgroups.target
label_names = newsgroups.target_names

print(f'Total documents: {len(documents)}')
print(f'Categories: {label_names}')
print(f'\nSample document:\n{"-"*50}\n{documents[0][:300]}...')
```

## 🔍 Step 3: Exploratory Data Analysis (EDA)

```python
# Distribution of documents per category
df = pd.DataFrame({'document': documents, 'label': true_labels})
df['category'] = df['label'].apply(lambda x: label_names[x])
df['doc_length'] = df['document'].apply(lambda x: len(x.split()))

print('Documents per category:')
print(df['category'].value_counts())
print(f'\nAverage document length: {df["doc_length"].mean():.0f} words')

# Plot distribution
plt.figure(figsize=(10, 4))
df['category'].value_counts().plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Document Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 🧹 Step 4: Text Preprocessing

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    """
    Full text preprocessing pipeline:
    1. Lowercase
    2. Remove special characters & numbers
    3. Remove stopwords
    4. Lemmatize words
    """
    # Lowercase
    text = text.lower()
    # Remove special characters, numbers, extra whitespace
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenize
    tokens = text.split()
    # Remove stopwords and short words
    tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
    # Lemmatize
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return ' '.join(tokens)

print('Preprocessing documents...')
cleaned_docs = [preprocess(doc) for doc in documents]
print(f'✅ Preprocessing complete!')
print(f'\nBefore: {documents[0][:200]}')
print(f'\nAfter: {cleaned_docs[0][:200]}')
```
